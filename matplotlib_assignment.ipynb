```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid()
plt.show()
```

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
```

```python
plt.bar(students,marks)
plt.show()
```

```python
colors = ['green','red','blue','orange']
plt.bar(students,marks,color=colors)
plt.title('Bar Chart')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.show()
```

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
```

```python
max_index = revenue.index(max(revenue))
explode = [0.1 if i == max_index else 0 for i in range(len(revenue))]
plt.pie(revenue,labels=regions,autopct='%1.1f%%',explode=explode,shadow= True)
plt.title('Companys revenue from different regions')
plt.show()
```

```python
random = np.random.randint(1,101,size=(101))
random
```
